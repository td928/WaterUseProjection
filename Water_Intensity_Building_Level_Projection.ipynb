```python
def clean_ll84(df, key):
    
    '''
    Clean the LL84 dataset by log transforming the dataset and removing any values 
    that is outliers in the dataset and return 
    '''
    
    std = df[key].std()
    mean = df[key].mean()
    
    df['Outliers'] = abs(df[key] - mean) > 2 * std
        
    df = df[df.Outliers == False]
    
    return df 
```
